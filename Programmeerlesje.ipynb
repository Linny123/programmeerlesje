```python
import openpyxl
import pandas as pd  
```

```python
movies = pd.read_excel('excelVoorLesje.xlsx')
```

```python
# Namen van de Kolommen?
print(movies.columns)

# Hoeveel films gezien?
gezien = movies[movies.seen == 1]
print(len(gezien))

# lijst met titels van films op alfabetische volgorde die ik nog moet zien
nietGezien = movies[movies.seen == 0]
FORMAT = ['Title', 'Rating']
movies_not_seen = nietGezien[FORMAT]
print(movies_not_seen.sort_values(['Title'], ascending=True) )

# niet gezien, meer dan 250.000 stemmen
veelstemmen = movies[(movies.votes >= 250000) & (movies.seen == 0)]
print(veelstemmen[FORMAT].sort_values(['Title'], ascending = True))

#Niet gezien, meer dan 250.000 stemmen, na 2008
VeelNaNiet = veelstemmen[(veelstemmen.Year > 2008)]
print(VeelNaNiet[FORMAT])

# Alle films die een 9.0 hebben gehaald
negen = movies[movies.Rating == 9.0]
print(negen[FORMAT])

#gemiddelde cijfer
print(movies.Rating.sum()/250)
```
